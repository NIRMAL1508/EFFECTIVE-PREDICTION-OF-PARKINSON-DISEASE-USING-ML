```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.dummy import DummyClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import tree
from sklearn import metrics
from sklearn.metrics import roc_curve,auc
from sklearn.model_selection import cross_val_predict
```

```python
X = parkinsons_data.drop(['status', 'name'], axis = 1)
Y = parkinsons_data.status
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X)
X = pd.DataFrame(sc.transform(X), columns = X.columns)
```

```python
fig, axes=plt.subplots(5,5,figsize=(15,15))
axes=axes.flatten()

for i in range(1,len(X.columns)):
    sns.boxplot(x=X[X.columns[i]],orient='v', ax=axes[i])
plt.tight_layout()
plt.show()
```

```python
X_train, X_test, y_train,  y_test = train_test_split(X, Y,train_size=0.80, test_size=0.20, random_state=42)
print(len(X_train)),print(len(X_test))
```

```python
score_liste=[]
precision_scor=[]
f1_scor = []
recall_scor=[]
sppt_scor = []
from sklearn.metrics import roc_curve
from sklearn.metrics import precision_recall_fscore_support as score
```

```python
def modeleff(y,y_head):
  precision,recall,fscore,support=score(y,y_head,average='macro')
  precision_scor.append(precision)
  recall_scor.append(recall)
  f1_scor.append(fscore)
  sppt_scor.append(support)
  score_liste.append(accuracy_score(y,y_head))
```

**Ensemble**

```python
from sklearn.ensemble import RandomForestClassifier, VotingClassifier

# Train multiple Random Forest classifiers with different hyperparameters
rf1 = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
rf2 = RandomForestClassifier(n_estimators=500, max_depth=10, random_state=42)
rf3 = RandomForestClassifier(n_estimators=100, max_depth=20, random_state=42)
rf4 = RandomForestClassifier(n_estimators=500, max_depth=20, random_state=42)

# Combine the classifiers using a VotingClassifier
ensemble = VotingClassifier(estimators=[('rf1', rf1), ('rf2', rf2), ('rf3', rf3), ('rf4', rf4)], voting='hard')
ensemble.fit(X_train, y_train)
y_pred = ensemble.predict(X_test)
# Evaluate the ensemble on the test set
print("Ensemble accuracy:", ensemble.score(X_test, y_test))
```

**HyperParameter**

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 500, 1000],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
rf = RandomForestClassifier()
# Perform a grid search to find the best hyperparameters
grid_search = GridSearchCV(rf, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
print(grid_search.best_params_)
# Train a Random Forest classifier with the best hyperparameters
best_rf = RandomForestClassifier(**grid_search.best_params_, random_state=42)
best_rf.fit(X_train, y_train)
y_pred = best_rf.predict(X_test)
# Evaluate the model on the test set
print("Random Forest accuracy:", best_rf.score(X_test, y_test))
```

**Feature Importances**


```python
from sklearn.feature_selection import SelectFromModel

# Define a Random Forest classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the Random Forest classifier
rf.fit(X_train, y_train)

# Print the feature importances
print("Feature Importances:", rf.feature_importances_)

# Use the SelectFromModel object to select the most important features
sfm = SelectFromModel(rf, threshold=0.05)
sfm.fit(X_train, y_train)

# Transform the training and testing sets to only use the selected features
X_train_selected = sfm.transform(X_train)
X_test_selected = sfm.transform(X_test)

# Train a new Random Forest classifier using the selected features
rf_selected = RandomForestClassifier(n_estimators=100, random_state=42)
rf_selected.fit(X_train_selected, y_train)

# Evaluate the model on the test set
print("Random Forest accuracy (all features):", rf.score(X_test, y_test))
print("Random Forest accuracy (selected features):", rf_selected.score(X_test_selected, y_test))
y_pred = rf_selected.predict(X_test_selected)
modeleff(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
